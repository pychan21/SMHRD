##### 3. 데이터 전처리

```python
import glob
import os
# 폴더 위치에 'winequality 의 이름이 있는 파일의 위치의 경로를 가져옴
path = os.path.join('winequality*')
```

```python
# 경로 내 파일을 리스트로 가져옴
all_file_names = glob.glob(path)
all_file_names
```

```python
# i값을 만들어 인덱스에 추가해줌
all_df_list = []
for i, file in enumerate(all_file_names) :
    df = pd.read_csv(file, sep = ";")
    df['type'] = i # 0 = red, 1 = white
    all_df_list.append(df)
```

```python
for file in all_file_names :
    df = pd.read_csv(file, sep = ";")
    display(df)
```

```python
# ignore_index=True 인덱스번호를 무시함
wine =pd.concat(all_df_list, ignore_index=True)
wine
```

```python
# 산점도와 회귀선
# 화이트 와인과 레드 와인을 몇 개만 샘플링 해서 시각화 하자
```

```python
def take_sample(df,n=250):
    return df.loc[np.random.choice(df.index,size = n, replace= False)] # 비복원추출
```

```python
red_samples = take_sample(wine[wine['type']==0])
```

```python
white_samples = take_sample(wine[wine['type']==1])
```

```python
white_samples.shape
```

```python
wine_samples = pd.concat([red_samples,white_samples])
```

```python
sns.pairplot(wine_samples,
            kind = 'reg',
            hue='type'
            )
```

```python
sns.pairplot(wine_samples,
            kind = 'reg',
            hue='type',
            vars=['alcohol','quality','residual sugar'])
```

#####  5. 모델 선택 및 하이퍼 파라미터 튜닝

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import LinearSVC
```

```python
logi_model = LogisticRegression()
svc_model = LinearSVC()
```

```python
from sklearn.model_selection import train_test_split
```

```python
wine.columns
```

```python
X = wine[['volatile acidity','total sulfur dioxide']]
Y = wine['type']
```

```python
X.shape ,Y.shape
```

```python
X_train, X_test, Y_train,Y_test = train_test_split(X,Y)
```

```python
from sklearn.model_selection import cross_val_score
```

```python
logi_result = cross_val_score(logi_model, X_train , Y_train , cv = 5)
logi_result.mean()
```

```python
svc_result = cross_val_score(svc_model, X_train , Y_train , cv = 5)
svc_result.mean()
```

```python
# 규제 값 튜닝,
C = [0.1, 1, 10, 100]
for i in C:
    svc_model = LinearSVC(C = i)

    svc_result = cross_val_score(svc_model, X_train, Y_train, cv=5)
    print(f'C:{i}, score:{svc_result.mean():.4}')
```
