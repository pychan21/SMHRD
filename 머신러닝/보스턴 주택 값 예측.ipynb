```python
from sklearn.datasets import load_boston #보스턴 주택 값 데이터 
```

```python
boston = load_boston()
```

```python
# boston 의 data 를 df 로 만들기
# boston 의 feature_name을  columns로 사용하기
```

```python
boston_df = pd.DataFrame(boston.data,
                        columns = boston.feature_names)
boston_df
```

```python
house_price=boston.target
```

##### train과 test로 분리

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Lasso,Ridge
```

```python
X_train, X_test, Y_train,Y_test = train_test_split(
boston_df,house_price)
```

```python
boston_copy = X_train.copy()
boston_copy['Price'] = Y_train
boston_copy
```

```python
boston_copy.corr()
```

#####  
특성 확장
칼럼값을 곱해서 새로운 칼럼을 만들자

```python
extend_X_train = X_train.copy()
for col1 in X_train.columns:
    for col2 in X_train.columns:
        extend_X_train[col1+"X"+col2] = X_train[col1]*X_train[col2]
```

```python
extend_X_test = X_test.copy()
for col1 in X_test.columns:
    for col2 in X_test.columns:
        extend_X_test[col1+"X"+col2] = X_test[col1]*X_test[col2]
```

```python
lasso = Lasso(alpha = 10)
lasso.fit(extend_X_train,Y_train)
print(lasso.score(extend_X_train,Y_train))
print(lasso.score(extend_X_test,Y_test))
print(np.sum(lasso.coef_!=0))
```
