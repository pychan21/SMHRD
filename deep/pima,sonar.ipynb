```python
import pandas as pd
# 피마 인디언 당뇨병 데이터
pima = pd.read_csv("pima.csv", header=None,
                  names=["pregnant", "plasma", "pressure", "thickness", "insulin", "BMI", "pedigree", "age", "class"])
pima.head()
```

```python
# 상관관계 분석 (seaborn 라이브러리를 이용해서 시각화)
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12,12))

sns.heatmap(pima.corr(),annot=True)
```

```python
# 신경망 설계
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

model=Sequential()
# 입력층 설계
model.add(Dense(24, input_dim=8, activation ="relu"))

# 은닉층 설계
model.add(Dense(16, activation ="relu"))
model.add(Dense(8, activation ="relu"))
model.add(Dense(4, activation ="relu"))
model.add(Dense(2, activation ="relu"))

# 출력층 설계
model.add(Dense(1,activation="sigmoid"))

model.summary()
```

```python
model.compile(loss="binary_crossentropy",
             optimizer="adam",
             metrics=["acc"])
```

```python
model.fit(pima.iloc[:,:-1],pima.iloc[:,-1],epochs=500, batch_size=42)
```

##### 베스트 모델 찾아서 저장 :  ModelCheckpoint
더 이상 학습이 나아지지 않으면 중단하는 :EarlyStopping


```python
import os
from tensorflow.keras.callbacks import ModelCheckpoint , EarlyStopping
#모델을 저장할 폴더명
model_folder = "./model/"
# 해당 폴더가 없다면
if not os.path.exists(model_folder):
    os.mkdir(model_folder)
    
#저장할 파일명 설정
#{epoch:04d} : 반복수를 4자리로 설정
#{acc:.4f} : 정확도를 소수점 4자리까지 표시

modelpath = model_folder + "pima-{epoch:04d}-{val_acc:.4f}.hdf5"

#베스트를 찾아서 해당파일명으로 저장
# save_best_only : 더 나은 결과값만 저장
#modelcheckpoint(filepath=파일패스 , monitor=기준값,save_vest_only=True/False)

# 매 에포크 마다의 훈련 손실값 (loss)
# 매 에포크 마다의 훈련 정확도 (acc)
# 매 에포크 마다의 검증 손실값 (val_loss)
# 매 에포크 마다의 검증 정확도 (val_acc)

mc = ModelCheckpoint(filepath=modelpath,
                    monitor="val_acc",
                    save_best_only=True,
                    verbose=1)

# EarlyStopping(monitor=기준값 ,patience= 조금 더 기다리는 횟수)
# patience=20 -> 학습이 더 나아지지 않더라도 20회는 더 반복해줌

es = EarlyStopping(monitor= "val_acc",
                  patience=100)
#학습
#validation_split = 0.33 -> 검증데이터를 0.33만큼 추출해서 평가
# 0.66 학습에 사용
# callbacks -> mc와 es 두개를 조합해서 검증을함
history = model.fit(pima.iloc[:,:-1],pima.iloc[:,-1],
                    epochs=50000, batch_size=50,
                   validation_split=0.33,
                   callbacks=[mc,es])
```

##### 결과 시각화

```python
import matplotlib.pyplot as plt
import numpy as np
acc = history.history["acc"]
val_acc = history.history["val_acc"]
acc_len = np.arange(len(acc))

plt.plot(acc_len,acc,c="red")
plt.plot(acc_len,val_acc,c ="blue")
```

```python
import pandas as pd

sonar = pd.read_csv("sonar.csv", header=None)

sonar.head()
# sonar -> 음파탐지기로 광석과 일반 돌을 측정한 데이터
```

```python
X = sonar.iloc[:,:-1]
y = sonar.iloc[:,-1]

X.shape,y.shape
```

```python
# y의 클래스 수
y.unique()
```

```python
y_en = pd.get_dummies(y)
y_en.shape
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y_en,
                                                   random_state=0)
```

```python
X_train.shape, X_test.shape, y_train.shape,y_test.shape
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout

model_sonar=Sequential()
# 입력층 설계
model_sonar.add(Dense(120, input_dim=60, activation ="relu"))

# 은닉층 설계
model_sonar.add(Dense(240, activation ="relu"))
model_sonar.add(Dense(120, activation ="relu"))
model_sonar.add(Dense(60, activation ="relu"))
model_sonar.add(Dense(30, activation ="relu"))
model_sonar.add(Dense(10, activation ="relu"))




# 출력층 설계
model_sonar.add(Dense(2,activation="softmax"))

model_sonar.summary()
```

```python
model_sonar.compile(loss="binary_crossentropy",
             optimizer="adam",
             metrics=["acc"])
```

```python
h2=model_sonar.fit(X_train,y_train,
                epochs=200, batch_size=5,
               validation_data=(X_test,y_test))
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout

model_sonar2=Sequential()
# 입력층 설계
model_sonar2.add(Dense(120, input_dim=60, activation ="relu"))
model_sonar2.add(Dropout(0.50))
# 은닉층 설계
model_sonar2.add(Dense(240, activation ="relu"))
model_sonar2.add(Dropout(0.50))

model_sonar2.add(Dense(120, activation ="relu"))
model_sonar2.add(Dropout(0.50))

model_sonar2.add(Dense(60, activation ="relu"))
model_sonar2.add(Dropout(0.50))

model_sonar2.add(Dense(30, activation ="relu"))
model_sonar2.add(Dropout(0.50))

model_sonar2.add(Dense(10, activation ="relu"))




# 출력층 설계
model_sonar2.add(Dense(2,activation="softmax"))

model_sonar2.summary()
```

```python
model_sonar2.compile(loss="binary_crossentropy",
             optimizer="adam",
             metrics=["acc"])
```

```python
h3=model_sonar2.fit(X_train,y_train,
                epochs=200, batch_size=5,
               validation_data=(X_test,y_test))
```

```python
acc = h2.history["acc"]
val_acc = h2.history["val_acc"]

acc_len = np.arange(len(acc))

plt.plot(acc_len, acc, c="red")
plt.plot(acc_len, val_acc, c="blue")
```

```python
acc = h3.history["acc"]
val_acc = h3.history["val_acc"]

acc_len = np.arange(len(acc))

plt.plot(acc_len, acc, c="red")
plt.plot(acc_len, val_acc, c="blue")
```

```python
print("훈련점수 : ",model_sonar.evaluate(X_train,y_train))
print("테스트점수 : ",model_sonar.evaluate(X_test,y_test))
```

```python
print("훈련점수",model_sonar2.evaluate(X_train,y_train))
print("테스트점수",model_sonar2.evaluate(X_test,y_test))
```
