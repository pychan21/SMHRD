#####  활성화 함수
- 입력에 일정 값을 곱해서 출력값을 결정하는 것 (실제 신경망으로 동작을 모방
- 종류 : 계단함수,tanh , sigmoid,ReLu 등

#####  계단함수
0보다 크면 1 이고 그렇지 않으면 0인 함수

```python
import matplotlib.pyplot as plt
import numpy as np
def stepFun(x):
    # x가 0보다 크면 1이 반환 ,그렇지 않으면 0이 반환
    return np.array(x >0, dtype = np.int)
x = np.arange(-5.0,5.0,0.1)
y = stepFun(x)

plt.plot(x,y)
```

```python
def sigmoidFun(x):
    return 1/(1+ np.exp(-x))
x = np.arange(-5.0,5.0,0.1)
y = sigmoidFun(x)

plt.plot(x,y)
```

#####  sigmoid 미분

```python
def d_sigmoidFun(x) :
    result = sigmoidFun(x) *(1 - sigmoidFun(x))
    
    return result
x = np.arange(-5.0,5.0,0.1)
y = d_sigmoidFun(x)

plt.plot(x,y)
```

```python
x = np.arange(-5.0,5.0,0.1)
y = np.tanh(x)

plt.plot(x,y)
```

```python
def d_tanhFun(x) :
    return (1+np.tanh(x)) *(1-np.tanh(x))
x = np.arange(-5.0,5.0,0.1)
y = d_tanhFun(x)

plt.plot(x,y)
```

##### ReLU 함수

```python
def reluFUhn(x):
    return np.maximum(0,x)
x = np.arange(-5.0,5.0,0.1)
y = reluFUhn(x)

plt.plot(x,y)
```
