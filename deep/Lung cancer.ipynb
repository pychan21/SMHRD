```python
# x,y로 분리
x = data.iloc[:,:-1]
y = data.iloc[:,-1]

x.shape, y.shape
```

```python
# w,b를초기화 하기 위한 seed 설정
import numpy as np
import tensorflow as tf
seed = 7 
np.random.seed(seed)
tf.random.set_seed(seed)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# 새로운 신경망 생성
model = Sequential()

#신경망에 입력층을 추가
#input_dim : 특징의 개수
model.add(Dense(34,input_dim=17 ,activation="relu"))
# 신경망에 출력층을 추가
# 회귀 : linear(기본값이라 생략가능)
# 이진분류 : 1 (sigmoid), 2(원핫인코딩,softmax)
# 다중분류 : softmax
model.add(Dense(1,activation="sigmoid"))

model.summary()
```

```python
# 컴파일
model.compile(loss="binary_crossentropy",
             optimizer="adam",
             metrics=["acc"])
```

```python
# 학습
model.fit(x,y, epochs=30, batch_size=10)
```

```python
model2.fit(x,y_en, epochs=200, batch_size=10)
```

```python
from sklearn.datasets import load_boston
from tensorflow.keras.datasets import boston_housing
```

```python
(x_train,y_train), (x_test,y_test) = boston_housing.load_data()
```

```python
x_train.shape, x_test.shape , y_train.shape, y_test.shape
```

```python
model3 = Sequential()

#입력층
model3.add(Dense(26, input_dim=13, activation ="relu"))
#은닉층
model3.add(Dense(13, activation ="relu"))

#출력층
model3.add(Dense(1,activation="linear"))

model3.summary()
```

```python
#컴파일
#mse 또는 mean squared error
#회귀에서 정확도가 필요없으므로 METRICS는 사용하지 않아도 상관없음 - loss만 평가
model3.compile(loss="mse",
              optimizer="adam")
```

```python
# verbose=0 : 실행과정을 보여주지 않음
model3.fit(x_train, y_train,epochs=200,batch_size=10, verbose=0)
```

```python
model3.evaluate(x_train,y_train)
```

##### BMI 데이터셋을 이용한 다중 분류 학습

```python
bmi = pd.read_csv("bmi.csv")
```

```python
bmi
```

```python
# label의 클래스마다의 데이터의 수
bmi.label.value_counts()
```

```python
# 딥러닝을 이용하여 다중분류
```

```python
x = bmi.iloc[:,:-1]
```

```python
x
```

```python
y = bmi["label"]
```

```python
Y = pd.get_dummies(y)
```

```python
Y.shape
```

```python
# 새로운 신경망 생성
model1 = Sequential()

#신경망에 입력층을 추가
#input_dim : 특징의 개수
model1.add(Dense(16,input_dim=2 ,activation="relu"))
# 신경망에 은닉충 추가
model1.add(Dense(8,activation="relu"))
model1.add(Dense(4,activation="relu"))

# 신경망에 출력층을 추가
# 회귀 : linear(기본값이라 생략가능)
# 이진분류 : 1 (sigmoid), 2(원핫인코딩,softmax)
# 다중분류 : softmax
model1.add(Dense(3,activation="softmax"))

model1.summary()
```

```python
# 컴파일
model1.compile(loss="categorical_crossentropy",
             optimizer="adam",
             metrics=["acc"])
```

```python
model1.fit(x,Y, epochs=100, batch_size=10)
```
