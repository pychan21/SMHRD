```python
# 데이터 로드
from sklearn.datasets import load_iris
iris = load_iris()
iris.keys()
```

```python
iris.data.shape
```

```python
iris.target_names
```

```python
#원핫 인코딩 (라벨인코딩 -> 원핫인코딩)
from sklearn.preprocessing import LabelEncoder
from tensorflow.keras.utils import to_categorical

e = LabelEncoder()
e.fit(iris.target)

y = e.transform(iris.target)
```

```python
y
```

```python
# 원핫 인코딩
y_en = to_categorical(y)
y_en.shape
```

```python
# 초기화를 위한 seed 설정
import numpy as np
import tensorflow as tf
#seed = 0
#np.random.seed(seed)
#tf.random.set_seed(seed)
```

```python
#신경망 설계
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

model = Sequential()
#입력층
model.add(Dense(40, input_dim=4, activation ="relu"))
#은닉층
model.add(Dense(28, activation ="relu"))
model.add(Dense(20, activation ="relu"))
model.add(Dense(12, activation ="relu"))


#출력층
model.add(Dense(3,activation="softmax"))

model.summary()
```

```python
model.compile(loss="categorical_crossentropy",
             optimizer="adam",
             metrics=["acc"])
```

```python
model.fit(iris.data,y_en, epochs=500, batch_size=10)
```

```python
model.save("model_iris.h5")
```

##### 딥러닝 교차검증

```python
from sklearn.model_selection import KFold # 몇개로 구분할지
from sklearn.model_selection import cross_val_score

def deep_model():
    model1 = Sequential()
    #입력층
    model1.add(Dense(40, input_dim=4, activation ="relu"))
    #은닉층
    model1.add(Dense(28, activation ="relu"))
    model1.add(Dense(20, activation ="relu"))
    model1.add(Dense(12, activation ="relu"))


    #출력층
    model1.add(Dense(3,activation="softmax"))

    model1.compile(loss="categorical_crossentropy",
                 optimizer="adam",
                 metrics=["acc"])

    return model1
```

```python
from tensorflow.keras.wrappers.scikit_learn import KerasClassifier
```

```python

model1 = KerasClassifier(build_fn=deep_model,
                       epochs=50, batch_size=10)
fold = KFold(n_splits=5, shuffle=True, random_state=0)
score = cross_val_score(model1,iris.data,y_en, cv =fold)
```

```python
import numpy as np
print(np.mean(score))
```
