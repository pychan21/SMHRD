##### 침식과 팽창

- 이진 이미지 처리 방법
- Erosion(침식) : 이미지 경계 부분에서 0을 1로 변환하는 방법
- Dilation (팽창) : 이미지 경계 부분에서 1을 0으로 변환하는 방법
- iteration. : 몇 픽셀 씩 변환할 것인지 설정

```python
import cv2
import numpy as np

# 흑백 이미지로 읽기
img = cv2.imread("333.png",cv2.IMREAD_GRAYSCALE)

cv2.imshow("orginal",img)

# 침식과 팽창의 크기를 설정
kernel = np.ones((3,3),np.uint8)

# 침식
erosion = cv2.erode(img, kernel, iterations=1)
cv2.imshow("erosion",erosion)

# 팽창
dilation = cv2.dilate(img,kernel, iterations=1)

cv2.imshow("dilation",dilation)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

#####  침식과 팽창을 이용해서 솔트 잡음을 제거
- Openning : 침식 후에 팽창 연상을 수행(검정 배경에 흰색 잡음이 있는 경우)
- Closing : 팽창 후에 침식 연산을 수행 (흰색 글씨에 검정색 잡음이 있는경우)

```python
# Opening
import cv2
import numpy as np 

img = cv2.imread("A.png", cv2.IMREAD_GRAYSCALE)

cv2.imshow("original",img)

#Opening
kernel = np.ones((3,3), np.uint8)

opening = cv2.morphologyEx(img,cv2.MORPH_OPEN, kernel)

cv2.imshow("opening",opening)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# closing
import cv2
import numpy as np 

img = cv2.imread("B.png", cv2.IMREAD_GRAYSCALE)

cv2.imshow("original",img)

#closing
kernel = np.ones((5,5), np.uint8)

closing = cv2.morphologyEx(img,cv2.MORPH_CLOSE, kernel)

cv2.imshow("closing",closing)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np
img1 = cv2.imread("game.png")
img2 = cv2.imread("apple.png",cv2.IMREAD_GRAYSCALE)

img1_gray = cv2.cvtColor(img1,cv2.COLOR_BGR2GRAY)

# 템플릿 이미지의 크기 계산
# [::-1] -> 첫번째 인덱스 값과 두 번째 인덱스 값을 서로 바꿈
w,h = img2.shape[::-1]

# 템플릿 매칭
res = cv2.matchTemplate(img1_gray,img2, cv2.TM_CCOEFF_NORMED)

# 매칭도가 기준값 이상인 위치만 가져온다
loc=np.where(res >=0.7)

for pt in zip(*loc[::-1]):
    cv2.rectangle(img1,pt,(pt[0]+w, pt[1]+h),(0,0,255),2)

cv2.imshow("match",img1)



cv2.waitKey(0)
cv2.destroyAllWindows()
```

#####  hough 변환을 이용해서 이미지에 원 찾기

```python
import cv2
import numpy as np

img = cv2.imread("555.jpg")
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# 허프 변환으로 원 찾기
# (이미지, 검색방법, 원의 중심점간의 최소거리, ?,
# param1=Canny 연산자의 최대값, param2=카운트값(작으면 많은 원을 찾음),
# minRadius=원의 최소 반지름", maxRadius =원의 최대 반지름)
circle = cv2.HoughCircles(img_gray, cv2.HOUGH_GRADIENT, 1, 10, 
                         param1=60,
                         param2=50,
                         minRadius=10,
                         maxRadius=100)

# 원이 검색되었다면
if circle is not None:
    # 반올림
    circle = np.uint16(np.around(circle))
    
    # 검색된 원을 그려준다
    for i in circle[0, :] :
        cv2.circle(img, (i[0], i[1]), i[2], (255, 0, 0), 4)
        
    cv2.imshow("circle", img)        
    
    cv2.waitKey(0)
    cv2.destroyAllWindows()
else :
    print("원이 없습니다")    
```

```python
import cv2
import numpy as np

img = cv2.imread("0000.jpg")
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# 허프 변환으로 원 찾기
# (이미지, 검색방법, 원의 중심점간의 최소거리, ?,
# param1=Canny 연산자의 최대값, param2=카운트값(작으면 많은 원을 찾음),
# minRadius=원의 최소 반지름", maxRadius =원의 최대 반지름)
# cv2.HoughCircles 뒤에 값을 늘려줄수록 원 사이의 크기가 멀어짐
circle = cv2.HoughCircles(img_gray, cv2.HOUGH_GRADIENT, 1, 100, 
                         param1=60,
                         param2=50,
                         minRadius=10,
                         maxRadius=100)

# 원이 검색되었다면
if circle is not None:
    # 반올림
    circle = np.uint16(np.around(circle))
    
    # 검색된 원을 그려준다
    for i in circle[0, :] :
        cv2.circle(img, (i[0], i[1]), i[2], (255, 0, 0), 4)
        
    cv2.imshow("circle", img)        
    
    cv2.waitKey(0)
    cv2.destroyAllWindows()
else :
    print("원이 없습니다")    
```

#####  모서리 (corner) 검출

```python
import cv2
import numpy as np

img = cv2.imread("999.png")
imgray = (cv2.cvtColor(img,cv2.COLOR_BGR2GRAY))

# (이미지 , 코너 갯수 , 기준값, 코너점 가느이 최소 거리)
corners = cv2.goodFeaturesToTrack(imgray,25,0.01,10)
#좌표값이므로 정수로 변환
corners = np.int0(corners)

# 코너 좌표에 점을 찍어준다
for i in corners:
    #ravel(): 다차원의 배열을 1차원 배열로 변환(리스트, 튜플, 배열 등 모두 사용)
    #flatten(): 다차원의 배열을 1차원 배열로 변환(배열만 사용 가능)
    x, y = i.ravel()
    # -1 : 원의 내부의 색을 채움
    cv2.circle(img,(x,y),3,(255,0,0),-1)
cv2.imshow("corner", img) 
cv2.waitKey(0)
cv2.destroyAllWindows() 
```

##### Harr cascade 분류기를 이용한 얼굴 영역 검출

```python
import cv2

# 얼굴 검출기
face_cascade = cv2.CascadeClassifier("haarcascade_frontalface_default.xml")
#얼굴 안에 있는 눈 검출기
eye_cascade = cv2.CascadeClassifier("haarcascade_eye.xml")

# 얼굴과 눈을 검출할 이미지 로드
img = cv2.imread("888.png")
imggray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
#얼굴 검출기 적용해서 얼굴 좌표를 가져옴(시작X좌표,시작Y좌표,길이 ,높이)
# 이미지 명, 스케일 값, 최소 이웃값
faces = face_cascade.detectMultiScale(imggray, 1.3,5)

# 찾은 얼굴 좌표를 이용해서 원 이미지에서 얼굴 ROI를 사격형으로 그려줌
for(x,y,w,h) in faces :
    cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),2)
    
cv2.imshow("face", img) 
cv2.waitKey(0)
cv2.destroyAllWindows() 
```

```python
import cv2

# 얼굴 검출기
face_cascade = cv2.CascadeClassifier("haarcascade_frontalface_default.xml")
#얼굴 안에 있는 눈 검출기
eye_cascade = cv2.CascadeClassifier("haarcascade_eye.xml")

# 얼굴과 눈을 검출할 이미지 로드
img = cv2.imread("888.png")
imggray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
#얼굴 부분에서 눈 영역을 찾아서 표시
# 얼굴 roi를 저장


faces = face_cascade.detectMultiScale(imggray, 1.3,5)



# 찾은 얼굴 좌표를 이용해서 원 이미지에서 얼굴 ROI를 사격형으로 그려줌
for(x,y,w,h) in faces :
    cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),2)
    
    roi_gray = imggray[y:y+h,x:x+w]
    roi_color = img[y:y+h,x:x+w]
    # 눈 영역을 검출
    eyes = eye_cascade.detectMultiScale(roi_gray, 1.1,2)
    for(ex,ey,ew,eh) in eyes:
        cv2.rectangle(roi_color,(ex,ey),(ex+ew,ey+eh),(0,0,255),2)
    
cv2.imshow("face", img) 
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import pafy

url = "https://www.youtube.com/watch?v=4OjuSyXhP8Y" # 유튜브 영상 url
vPafy = pafy.new(url)
play = vPafy.getbest(preftype="mp4")

# 얼굴 검출기
face_cascade = cv2.CascadeClassifier("haarcascade_frontalface_default.xml")
#얼굴 안에 있는 눈 검출기
eye_cascade = cv2.CascadeClassifier("haarcascade_eye.xml")

# 카메라 연결
try:
    print("카메라 연결 성공")
    cap = cv2.VideoCapture(play.url)
except:
    print("카메라 연결 실패")
while True:
    ret, frame = cap.read()
    # 동영상에서 한 장의 이미지를 읽는 것을 실패했다면,
    if not ret:
        cap.release()
        cv2.destroyAllWindows()
        break

   
    
    imggray = cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    #얼굴 부분에서 눈 영역을 찾아서 표시
    # 얼굴 roi를 저장


    faces = face_cascade.detectMultiScale(imggray, 1.3,5)



    # 찾은 얼굴 좌표를 이용해서 원 이미지에서 얼굴 ROI를 사격형으로 그려줌
    for(x,y,w,h) in faces :
        cv2.rectangle(frame,(x,y),(x+w,y+h),(255,0,0),2)

        roi_gray = imggray[y:y+h,x:x+w]
        roi_color = frame[y:y+h,x:x+w]
        # 눈 영역을 검출
        eyes = eye_cascade.detectMultiScale(roi_gray, 1.1,2)
        for(ex,ey,ew,eh) in eyes:
            cv2.rectangle(roi_color,(ex,ey),(ex+ew,ey+eh),(0,0,255),2)

    cv2.imshow("face", frame) 
    k= cv2.waitKey(33)
    if k == 49: #숫자 1을 누른 경우
        cap.release()
        cv2.destroyAllWindows()
        break
```
