(5) 신경망 설계

- 파라미터 초기화 (w, b)를 위한 seed 설정
- 항상 같은 초기화값이 들어가도록 해야함

```python
import numpy as np
import tensorflow as tf

seed = 0
np.random.seed(seed)
tf.random.set_seed(seed)
```

- 신경망 설계

  - Sequential() : 신경망을 생성
  - Dense() : 신경망의 레이어를 설정
  - Activation() : 활성화 함수를 설정

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation
```

```python
model = Sequential()

# add() : 신경망에 레이어을 추가
# Dense(출력갯수, input_dim=입력갯수)
model.add(Dense(1, input_dim=2))
# Activation("활성화함수 종류 (sigmoid, tanh, relu, linear, softmax 등)")
model.add(Activation("sigmoid"))

# 설계한 신경망을 확인
model.summary()
```

- 모델 컴파일

```python
# compile(loss="", optimizer="", metrics=[""])
# loss : 손실함수 (mse 또는 mean sequared error(회귀), 
#                  binary_crossentropy(이진분류),
#                  categrical_crossentropy(다중분류)
# optimizer : 경사하강법의 종류 (rmpsprop, adam)
# metrics : 평가방법 (acc 또는 accuracy)
model.compile(loss="binary_crossentropy",
              optimizer="adam",
              metrics=["acc"])
```

(6) 학습

```python
# fit(특성데이터, 라벨데이터, epochs=값, batch_size=값)
# epochs : 오차 역전파를 몇 번 반복할 것인지 설정
# batch_size : 학습할  몇 개 데이터씩 사용할 것인지 설정

model.fit(XData, y, epochs=2000, batch_size=4)
```

신경망으로 직접 논리 게이트를 구현

```python
# AND 
def ANDGate(x1, x2):
    # 파라미터 초기화
    w1, w2, acti = 0.2, 0.2, 0.4
    
    y = w1 * x1 + w2 * x2
    
    if y < acti:
        return 0
    elif y >= acti:
        return 1

print(ANDGate(0, 0))    
print(ANDGate(0, 1)) 
print(ANDGate(1, 0)) 
print(ANDGate(1, 1))   

# 0, 0, 0, 1
```

```python
def NANDGate(x1, x2):
    # 파라미터 초기화
    w1, w2, acti = -0.2, -0.2, -0.3
    
    y = w1 * x1 + w2 * x2
    
    if y < acti:
        return 0
    elif y >= acti:
        return 1

print(NANDGate(0, 0))    
print(NANDGate(0, 1)) 
print(NANDGate(1, 0)) 
print(NANDGate(1, 1))  

# 1, 1, 1, 0
```

```python
def ORGate(x1, x2):
    # 파라미터 초기화
    w1, w2, acti = 0.7, 0.7, 0.3
    
    y = w1 * x1 + w2 * x2
    
    if y < acti:
        return 0
    elif y >= acti:
        return 1

print(ORGate(0, 0))    
print(ORGate(0, 1)) 
print(ORGate(1, 0)) 
print(ORGate(1, 1))  

# 0, 1, 1, 1
```

```python
def XORGate(x1, x2):
    s1 = NANDGate(x1, x2)
    s2 = ORGate(x1, x2)
    y = ANDGate(s1, s2)
    
    return y

print(XORGate(0, 0))    
print(XORGate(0, 1)) 
print(XORGate(1, 0)) 
print(XORGate(1, 1))  

# 0, 1, 1, 0
```

오차 역전파를 직접 구현

```python
import numpy as np

# 예측할 값, 반복횟수, 학습률
# 학습률 : w, b를 업데이트할 때 얼마만큼 크기만큼씩 업데이트할 것인지 결정
def backpropagation(X, epochs, lr):
    # w, b 초기화
    w = np.random.rand(1)
    b = np.random.rand(1)
    
    # 학습
    for i in range(epochs) :
        # 예측
        pred_y = w * X + b
        
        # 오차계산
        error = X - pred_y
     
        # 중간값 출력
        print("반복수 : {}, 예측값 : {}, 오차 : {}".format(i, pred_y, error))
              
        # 오차역전파 (w, b 값을 갱신)    
        # 미분 (기울기) 만큼 갱신 (비용함수)
        d1 = error / pred_y
        
        # w와 b를 w + lr * 비용함수
        w = w + lr * d1
        b = b + lr * d1

backpropagation(1, 100, 0.01)        
```

```python
import numpy as np

# 2개의 노드로 구성된 2층 신경망 
def backpropagation2(X, epochs, lr):
  # w와 b 초기화
  w1 = np.random.rand(1)
  b1 = np.random.rand(1)
  w2 = np.random.rand(1)
  b2 = np.random.rand(1)

  for i in range(epochs):
    # 층별 중간 값 계산
    pred_y1 = w1 * X + b1
    pred_y2 = pred_y1 * (w2 * X + b2)

    # 2층 오차 계산
    error = X - pred_y2

    print("반복수 : {}, 예측값 : {}, 오차 : {}".format(i, pred_y2, error))

    # 2층 오차 역전파
    # 비용함수 - error / pred_y
    d2 = error / pred_y2

    # w를 갱신 w + lr * 비용함수
    w2 = w2 + lr * d2
    b2 = b2 + lr * d2    

    # 1층 오차 역전파 (1층의 비용함수와 2층의 비용함수를 곱함)
    d1 = pred_y2 / pred_y1 * d2

    # 오차 계산
    w1 = w1 + lr * d1
    b1 = b1 + lr * d1   

backpropagation2(1, 10, 0.01)
```
