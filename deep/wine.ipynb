```python
import pandas as pd
```

```python
wine = pd.read_csv("wine.csv")
```

```python
import numpy as np
import tensorflow as tf
seed = 7 
np.random.seed(seed)
tf.random.set_seed(seed)
```

```python
wine
```

```python
wine.shape
```

```python
# 데이터가 많을 경우 일부 데이터만 샘플링해서 사용
wine = wine.sample(frac=0.6) # 전체 데이터 중에서 60%만 사용
wine.shape
```

```python
# X,y 분리하기

# 회귀용 데이터 X1, y1 : y1 - alcohol
X1 = wine.drop('alcohol',axis=1)
y1 = wine['alcohol']

# 이진분류 데이터 X2, y2 : y2 - type
X2 = wine.drop('type',axis=1)
y2 = wine['type']

# 다중분류 데이터 X3, y3 : y3 - quality
X3 = wine.drop('quality', axis=1)
y3 = wine['quality']
```

```python
from sklearn.model_selection import train_test_split
y3_en = pd.get_dummies(y3)
y3_en
```

```python
X3_train, X3_test, y3_train, y3_test = train_test_split(X3,y3_en,random_state=0)
X3_train.shape, X3_test.shape, y3_train.shape, y3_test.shape
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
model1 = Sequential()

#입력층
model1.add(Dense(24, input_dim=12, activation ="relu"))
#은닉층
model1.add(Dense(36, activation ="relu"))
model1.add(Dense(24, activation ="relu"))
model1.add(Dense(14, activation ="relu"))







#출력층
model1.add(Dense(7,activation="softmax"))

model1.summary()
```

```python
model1.compile(loss="categorical_crossentropy",
              optimizer="adam",
              metrics=["acc"])
```

```python
model1.fit(X3_train,y3_train,epochs=1000,batch_size=20)
```

```python
# 예측
pred = model1.predict(X3_train)

for i in range(5):
    print("실제값 : {} \t 예측값 : {}".format(y3_train.iloc[i],pred[i]))
```
