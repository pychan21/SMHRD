```python
import pandas as pd
```

```python
lotto123 = pd.read_csv("lotto2.csv")
```

```python
lotto123
```

```python
# 보너스 번호 삭제
del lotto123["No_Bonus"]

lotto123.tail(10)
```

```python
X= lotto123.iloc[:,:-1]
y= lotto123.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,
                                                   random_state=45)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
y_tarin_en = pd.get_dummies(y_train)
y_test_en = pd.get_dummies(y_test)
```

```python
y_tarin_en.shape , y_test_en.shape
```

```python
y_tarin_en
```

#####  시드 값 초기화

```python
import numpy as np
import tensorflow as tf

seed = 0
np.random.seed(seed)
tf.random.set_seed(seed)
```

#####  모델 생성

```python
# 딥러닝 구동시 필요한 케라스 함수 호출
# Sequential -> 케라스에서 층을 쌓기 위해 사용하는 함수
# Dense -> 실제 신경망 층을 설정하는 기능
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense , Dropout

# Dropout -> 과적합 방지를 위해 사용함
# model.add(Dropout(0.50)) -> model에 데이터중 50%를 랜덤하게 날리고 ,
#살아있는 데이터에 가중치를 줘서 날린 데이터만큼 만들어줌

model = Sequential()
# 입력층
# add -> 층 추가
# activation -> 활성화 함수를 설정하는 기능(대부분 relu 사용)
# sigmoid -> 이진분류(1아니면 0)
# input_dim -> 입력할 노드 수(특성수)
model.add(Dense(81, input_dim=9 , activation ="relu"))
model.add(Dropout(0.50))
# 은닉층
# 은닉층은 사용자가 직접 설계를 해줄 수 있음
model.add(Dense(162, activation ="relu"))
model.add(Dropout(0.50))
model.add(Dense(81, activation ="relu"))
model.add(Dropout(0.50))
model.add(Dense(27, activation ="relu"))
model.add(Dropout(0.50))

# 출력층
# 회귀 -> linear(기본값이라 생략가능)
# 이진분류 -> sigmoid
# 원핫 인코딩을 했다면 softmax를 사용해야함
# 다중분류 -> softmax(각각 확률값으로 나옴 총합은 1)
model.add(Dense(3,activation="softmax"))

model.summary()
```

```python
model.compile(loss="categorical_crossentropy",
             optimizer="adam",
             metrics=["acc"])
```

```python
model.fit(X_train,y_tarin_en, epochs=10, batch_size=6)
```

```python
print("훈련 정확도 : ", model.evaluate(X_train,y_tarin_en))
print("테스트 정확도 : ", model.evaluate(X_test,y_test_en))
```
