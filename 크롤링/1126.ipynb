### 네이버 뮤직  TOP 10가져오기

```python
import requests as req
from bs4 import BeautifulSoup as bs
```

```python
res=req.get('https://music.naver.com/listen/top100.nhn?domain=TOTAL_V2')
```

```python
music = res.text
```

```python
music
```

```python
music_info = bs(music,'lxml')
```

```python
music_tex = music_info.find_all('td',class_ = 'ranking')
```

```python
rank=music_info.select('td.ranking')
```

```python
rank
```

```python
# _title과 title은 서로 다른 클래스명임 그렇기 때문에 가운데에 .으로 구분해줘야함
title = music_info.select('a._title.title')
```

```python
artist=music_info.select('td._artist > a')
```

```python
artist[0].text
```

```python
for i in rank :
    print(i.text)
```

```python
##필요없는 이스케이프코드를 지워주는 기능 >> strip()
artist[1].text.strip()
```

```python
import pandas as pd
```

```python
len(artist)
```

```python
rank_list = []
title_list = []
artist_list = []

for i in range(len(rank)) :
    rank_list.append(rank[i].text.strip())
    title_list.append(title[i].text.strip())
    artist_list.append(artist[i].text.strip())
    
```

```python
print(title_list)
```

```python
info = {"rank":rank_list,"title":title_list, "aritist":artist_list}
```

```python
movie_pd = pd.DataFrame(info)
```

```python
movie_pd
```

```python
movie_pd=movie_pd.set_index('rank')
```

```python
movie_pd.to_csv
```

```python
movie_pd.to_csv("노래순위.csv",encoding='euc-kr')
```
